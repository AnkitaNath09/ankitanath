### Correlation 

It tells about the relationship between quantitaive variables. The value oof correlation varies between -1 and 1 .
Positive or Negative SIgn Shows the direction of relation and strength varies between 0 & 1. 
<br>Weak  < 0.5
<br>Medium  - 0.5 to 8
<br>Strong  - 8 t0 ~ 1
<br> V.Strong  = 1



```python
#sns.get_dataset_names()
df=sns.load_dataset('diamonds')

df.describe(include=np.number)
```

```python
# this code will give the value of correlation between all quantitative variable
df=df[['carat','depth','table','price','x','y','z']]

df.corr()
```

```python
df_numeric=df.select_dtypes(include=[np.number])
```

```python
#is a measure of linear correlation between two sets of data.
#It is the covariance of two variables, divided by the product of their standard deviations; thus it is essentially a normalised measurement of the covariance,
#such that the result always has a value between −1 and 1
df_numeric.corr(method = 'pearson')
```

```python
#It is a measure of rank correlation: the similarity of the orderings of the data when ranked by each of the quantities.
#It is named after Maurice Kendall, who developed it in 1938,
    
df.corr(method = 'kendall')
```

```python
#Spearman's rank correlation coefficient or Spearman's ρ, named after Charles Spearman and 
#often denoted by the Greek letter {\displaystyle \rho }\rho  (rho) or as {\displaystyle r_{s}}r_{s}, is a nonparametric measure of rank correlation (statistical dependence between the rankings of two variables). 


cor=df.corr(method = 'spearman');cor
```

### Scatter Plots - Comapre the scatter plots with out tranformation and with log transformation

A scatter plot is a type of plot that shows the data as a collection of points. The position of a point depends on its two-dimensional value, where each value is a position on either the horizontal or vertical dimension

```python
# Code to plot a scatter plot in python
# Plot
plt.scatter(df['carat'] ,df['price'],color='orange', alpha=0.5)
plt.title('Scatter plot Carat and Price of Diamond')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
#Scatter plot with log trasformation

plt.scatter(np.log(df['carat']) ,np.log(df['price']),color='orange', alpha=0.5)
plt.xlabel('Carat')
plt.ylabel('Price')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```
