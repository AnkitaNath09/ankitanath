# UNIVARIATE

```python
#NAME_INCOME_TYPE vs Gender

plt.figure(figsize=[15,7])
sns.set_style('whitegrid')

sns.countplot(data=target_0, x='NAME_INCOME_TYPE', hue= 'CODE_GENDER', palette='copper')

plt.xticks(rotation=90)
plt.title('NAME_INCOME_TYPE vs Gender \n', fontsize=20)
plt.xlabel('INCOME TYPE',fontsize=15)
plt.ylabel('Gender',fontsize=20)
plt.yscale('log')

plt.show()
```

```python
#NAME_CONTRACT_TYPE vs Gender

plt.figure(figsize=[15,7])
sns.set_style('whitegrid')

sns.countplot(data=target_0, x='NAME_CONTRACT_TYPE', hue= 'CODE_GENDER', palette='muted')

plt.xticks(rotation=90)
plt.title('NAME_CONTRACT_TYPE vs Gender \n', fontsize=20)
plt.xlabel('INCOME TYPE',fontsize=15)
plt.ylabel('Gender',fontsize=20)
plt.yscale('log')

plt.show()
```

```python
#ORGANIZATION_TYPE vs Count

plt.figure(figsize=[15,30])
sns.set_style('whitegrid')

sns.countplot(data=target_0, y='ORGANIZATION_TYPE', order=target_0['ORGANIZATION_TYPE'].value_counts().index, palette='copper')

plt.xticks(rotation=90)
plt.yticks(fontsize=15)
plt.title('ORGANIZATION_TYPE', fontsize=20)
plt.xlabel('COUNT',fontsize=15)
plt.ylabel('ORGANIZATION_TYPE',fontsize=20)
plt.xscale('log')

plt.show()
```
